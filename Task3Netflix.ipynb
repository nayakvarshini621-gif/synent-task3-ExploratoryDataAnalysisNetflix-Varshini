```python
print("       Netflix Dataset Exploratory Data Analysis      ")
print("Project by : Varshini Nayak")
print("Tools used : Python , Pandas , Numpy , Matplotlib , Seaborn")
print("The Netflix dataset contains information about movies and TV shows available on Netflix includig genre , ratings , release data ,country and duration.")
print("This project performs exploratory data analysis (EDA) to identify trends , patterns, and insights from the dataset using python visualization libraries.")
print("Objective : ")
print("~ Analyze Netflix content distribution")
print("~ Identify popular genres and ratings")
print("~ Study contry-wise content production")
print("~ Understand trends in Netflix releases")
print("~ Visualize insights using graphs")
from datetime import date
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('netflix_titles.csv')
df.head()
df.info()
df.describe()
df.shape
df.isnull().sum()
plt.style.use('dark_background')
text_colums=['director' , 'cast', 'country' , 'rating']
for col in text_colums:
  df[col] = df[col].fillna("Unknown")
df.isnull().sum()
df.dtypes
df['date_added']=pd.to_datetime(df['date_added'] , errors='coerce')
df.describe(include='object')
numeric_df=df.select_dtypes(include=np.number)
corr=numeric_df.corr()
sns.heatmap(corr , annot=True , cmap='coolwarm')
plt.show()
print("The dataset contains limited numerical features , resulting in weak correlations overall.")
sns.countplot(x='type' , data=df , color="red")
plt.title("Movies vs TV Shows")
plt.show()
print("   ")
df['year_added']= df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='line',figsize=(12,6), color="red")
sns.set_style("darkgrid")
plt.title("Content Added Over Years ")
plt.xlabel("Years")
plt.ylabel( "Number of Movies")
plt.show()
print("Netflix content additions increased rapidly after 2015")
print("     ")
top_countries=df['country'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.set_style("darkgrid")
bars=plt.bar(
    top_countries.index,
    top_countries.values,
    color='#E50914'
    )
plt.title("Top 10 Countries Producing Netflix Content " , fontsize=16 , fontweight='bold')
plt.xlabel("Countries" , fontsize=12)
plt.ylabel("Number of Movies" , fontsize=12)
#Rotate country names
plt.xticks(rotation=45)
#Adding values on top of Bars
for bar in bars:
    yval=bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        yval + 5,
        int(yval),
        ha='center',
        fontsize=10
        )
plt.style.use('dark_background')
plt.grid(axis='y' , linestyle='--' , alpha=0.5) #add grid
plt.tight_layout() #adjust layout
plt.show()
print("USA contributes the highest amount of Netflix content , followed by India and the United Kingdom.")
print("     ")
#top 10 genres
top_genres=df['listed_in'].value_counts().head(10)
plt.figure(figsize=(14,8))
ax=plt.gca()
ax.set_facecolor('#141414')
plt.gcf().patch.set_facecolor('#141414')
sns.barplot(
    y=top_genres.index,
    x=top_genres.values,
    palette='Reds_r',
    hue=top_genres.index,
    legend=False
    )
plt.title(
    "Top 10 Most Popular Netflix Genres",
    fontsize=18,
    color='white',
    fontweight='bold',
    pad=20,
    )
plt.xlabel(
    "Number of Titles",
    fontsize=12,
    color='white'
)
plt.ylabel(
    "Genres",
    fontsize=12,
    color='white'
)
#tick colours
plt.xticks(color='white' , fontsize=10)
plt.yticks(color='white' , fontsize=10)
#Value Labels
for index,value in enumerate(top_genres.values):
    plt.text(
        value +5,
        index,
        str(value),
        color='white',
        va='center',
        fontsize=10
    )
#grid styling
plt.grid(
    axis='x',
    linestyle='--',
    alpha=0.3
)
plt.grid(axis='y' , linestyle='--' , alpha=0.5) #add grid
#sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
print(" International Movies and Dramas dominate the platform , showing Netflix's emphasis on globally diverse entertainment.")
print("      ")
plt.figure(figsize=(10,6))
plt.style.use('dark_background')

order=df['rating'].value_counts().head(10)
plt.figure(figsize=(12,6))
bars=plt.bar(
    order.index,
    order.values,
    color='#E50914'
    )
plt.title("Netflix Rating Distribution " , fontsize=16 , fontweight='bold')
plt.xlabel("Ratings" , fontsize=12)
plt.ylabel("Counts" , fontsize=12)
#Rotate country names
plt.xticks(rotation=45)
#Adding values on top of Bars
for bar in bars:
    yval=bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        yval + 5,
        int(yval),
        ha='center',
        fontsize=10
        )
plt.grid(axis='y' , linestyle='--' , alpha=0.5) #add grid
plt.tight_layout() #adjust layout
plt.show()
print("TV-MA and TV-14 are the m ost common ratings , indicating Netflix focuses heavily on mature and teenage audience. ")
print("      ")
print("KEY FINDINGS:")
print("Movies dominate Netflix compared to TV Shows")
print("USA is leading content producer")
print("International Movies and Dramas dominate the platform")
print("Mature audience ratings are most common")
print("Netflix content additions increased significantly after 2015")
print("      ")

print("CONCLUSION:")
print("This project successfully analyzed the Netflix Dataset using EDA techniques. Various trends related to genres,ratings , countries , and content growth were identified through visualizations.")
print("      ")
print("The project improved understanding of data cleaning and visualization , and insight generation using Python.")
```
